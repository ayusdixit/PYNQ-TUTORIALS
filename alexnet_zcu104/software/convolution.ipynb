```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

def conv2d(input, weight, bias, stride, padding):
    n, c, h, w = input.shape
    out_c, _, kh, kw = weight.shape

    # Apply padding
    padded = np.pad(input, ((0, 0), (0, 0), (padding, padding), (padding, padding)), mode='constant')

    out_h = (h + 2 * padding - kh) // stride + 1
    out_w = (w + 2 * padding - kw) // stride + 1

    output = np.zeros((n, out_c, out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):
            input_slice = padded[:, :, i * stride:i * stride + kh, j * stride:j * stride + kw]
            for k in range(out_c):
                output[:, k, i, j] = np.sum(input_slice * weight[k], axis=(1, 2, 3)) + bias[k]

    return output

def max_pool2d(input, kernel_size, stride):
    n, c, h, w = input.shape
    out_h = (h - kernel_size) // stride + 1
    out_w = (w - kernel_size) // stride + 1

    output = np.zeros((n, c, out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):
            output[:, :, i, j] = np.max(input[:, :, i * stride:i * stride + kernel_size, j * stride:j * stride + kernel_size], axis=(2, 3))

    return output

def adaptive_avg_pool2d(input, output_size):
    n, c, h, w = input.shape
    out_h, out_w = output_size

    stride_h, stride_w = h // out_h, w // out_w
    kernel_h, kernel_w = h - (out_h - 1) * stride_h, w - (out_w - 1) * stride_w

    output = np.zeros((n, c, out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):
            output[:, :, i, j] = np.mean(input[:, :, i * stride_h:i * stride_h + kernel_h, j * stride_w:j * stride_w + kernel_w], axis=(2, 3))

    return output

def numpy_inference(input_data, weights):
    # Features
    x = input_data

    x = conv2d(x, weights['features.0.weight'], weights['features.0.bias'], stride=4, padding=2)
    print("After conv1:", x.shape)
    x = relu(x)
    print("After rellu 1:", x.shape)
    x = max_pool2d(x, kernel_size=3, stride=2)
    print("After maxp1 1:", x.shape)

    # Conv2d -> ReLU -> MaxPool2d
    x = conv2d(x, weights['features.3.weight'], weights['features.3.bias'], stride=1, padding=2)
    print("After Conv2 :", x.shape)
    x = relu(x)
    print("After relu2:", x.shape)
    x = max_pool2d(x, kernel_size=3, stride=2)
    print("After maxpool 2:", x.shape)

    # # Conv2d -> ReLU
    x = conv2d(x, weights['features.6.weight'], weights['features.6.bias'], stride=1, padding=1)
    print("After conv3 :", x.shape)
    x = relu(x)
    print("After relu3:", x.shape)
    # # Conv2d -> ReLU
    x = conv2d(x, weights['features.8.weight'], weights['features.8.bias'], stride=1, padding=1)
    print("After Conv4:", x.shape)
    x = relu(x)
    print("After relu4 :", x.shape)
    # # Conv2d -> ReLU -> MaxPool2d
    x = conv2d(x, weights['features.10.weight'], weights['features.10.bias'], stride=1, padding=1)
    print("After Conv5:", x.shape)
    x = relu(x)
    print("After relu5:", x.shape)
    x = max_pool2d(x, kernel_size=3, stride=2)
    print("After maxpool5:", x.shape)

    # AdaptiveAvgPool2d
    x = adaptive_avg_pool2d(x, output_size=(6, 6))
    print("After addaptive  pool:", x.shape)

    # Flatten
    x = x.reshape(x.shape[0], -1)

    
    x = np.dot(x, weights['classifier.1.weight'].T) + weights['classifier.1.bias']
    print("After dense 1:", x.shape)
    x = relu(x)

    x = np.dot(x, weights['classifier.4.weight'].T) + weights['classifier.4.bias']
    print("After ddesne2 :", x.shape)
    x = relu(x)

    x = np.dot(x, weights['classifier.6.weight'].T) + weights['classifier.6.bias']
    print("After dnese 3:", x.shape)
   #Final softmax
    return x    #  have changed theis code ***************** softmax(x)

# Load the saved weights
saved_weights = np.load("../softwaree/model_weights.npy", allow_pickle=True).item()

# Prepare input data (assuming we have a sample image)
# Note: You'll need to preprocess your input data similarly to how it was done in the PyTorch model
sample_input = np.random.rand(1, 3, 214, 224).astype(np.float32)  # Replace with actual preprocessed image data

# Perform numpy inference
numpy_output = numpy_inference(sample_input, saved_weights)

print("Numpy inference output shape:", numpy_output.shape)
print("Numpy inference output:", numpy_output)

# Predictions
numpy_pred = np.argmax(numpy_output)

print("\nNumpy prediction:", numpy_pred)
```
